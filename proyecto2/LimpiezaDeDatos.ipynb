```python
#Importar librerias
import pandas as pd
```

```python
# Lectura de los datos
df_original = pd.read_csv("data.csv", delimiter=',')
```

```python
#Resumen del contenido de los datos
df_original.head()
```

```python
# Revisar las caracterÃ­sticas de los datos: completitud y tipo de dato
df_original.info()
```

```python
df_original = df_original[~df_original['EDUCATION'].isin([0, 5, 6])]

# Mostrar el DataFrame despuÃ©s de filtrar
print(df_original)
```

Como no hay datos nulos no hace falta limpiar la base de datos. Ya que no hay variables categoricas, no van a haber variables dummies y se lleva a cabo el debido analisis

```python
df_original = df_original[df_original['MARRIAGE'] != 0]

# Display the DataFrame after filtering
print(df_original)
```

```python
# Mapping dictionary for conversion
gender_mapping = {2: 0, 1: 1}

# Applying the mapping to the gender column
df_original['SEX'] = df_original['SEX'].map(gender_mapping)

# Displaying the DataFrame after conversion
print(df_original)
```

```python
# Codificar la variable education

dummy_quarter = pd.get_dummies(df_original['EDUCATION'], prefix='').astype(int)

# Drop the original 'education' column from df_clean
df_original.drop(['EDUCATION'], axis=1, inplace=True)

# Concatenate the original DataFrame with the dummy variables DataFrame
df_original = pd.concat([df_original, dummy_quarter], axis=1)

# Print the encoded DataFrame
df_original.info()
```

```python
# Codificar la variable day

dummy_casamiento = pd.get_dummies(df_original['MARRIAGE'], prefix='MARRIAGE').astype(int)

# Drop the original 'quarter' column from df_clean
df_original.drop(['MARRIAGE'], axis=1, inplace=True)

# Concatenate the original DataFrame with the dummy variables DataFrame
df_original = pd.concat([df_original, dummy_casamiento], axis=1)

# Print the encoded DataFrame
df_original.info()
```
